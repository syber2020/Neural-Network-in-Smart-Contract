```python
#Independent variables
import random
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('iris.csv',sep = ',')
df = df.replace('Iris-setosa', 1)
df = df.replace('Iris-versicolor', 2)
df = df.replace('Iris-virginica', 3)
# df["f1"] = 10 * df["f1"]
# df["f2"] = 10 * df["f2"]
# df["f3"] = 10 * df["f3"]
# df["f4"] = 10 * df["f4"]
labels = df['labels'].to_numpy()
df = df.drop(columns=['labels'])
input_set = df.to_numpy()
print(input_set)
```

```python
#Define Hyperparameters
np.random.seed(42)
weights = np.random.rand(4,1)
bias = np.random.rand(1)
lr = 0.05 #learning rate
print(input_set.shape)
print(weights.shape)
print(bias)
print(lr)
```

```python
#define activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#define derivative
def sigmoid_derivative(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
#train ANN model
for epoch in range(25000):
    inputs = input_set  # set aside the original input
    XW = np.dot(inputs, weights)+ bias  # dot product of input and add bias
    z = sigmoid(XW)  #sigmoid activation function
    
#     print ("sigmoid output is:",z )
    error = z - labels  #backprop step1
    
#     print ("error is ",error)
#     print("error sum is",error.sum())   #backpro step 2
    
    dcost = error
    dpred = sigmoid_derivative(z)
    
#     print("sigmoid derivative is ",dpred)
    z_del = dcost * dpred
    inputs = input_set.T 
#     print(inputs.shape)
#     print(z_del.shape)
#     print(weights.shape)  
    weights = weights - lr*np.dot(inputs, z_del)
    for num in z_del:
        bias = bias - lr*num
```

```python
print(weights)
print(bias)
```

# Predict Points

```python
single_pt = np.array([1,0,0])
print(np.dot(single_pt, weights))
result = sigmoid(np.dot(single_pt, weights) + bias)
print(result)
```
