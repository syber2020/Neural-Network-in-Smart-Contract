```python
#Independent variables
import random
import numpy as np
import pandas as pd
from numpy import loadtxt
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
dataset = loadtxt('pima-diabetes.txt', delimiter=',')
X = dataset[:500,0:8]
X1 = dataset[500:,0:8]
y = dataset[:500,8]
y1 = dataset[500:,8]
type(X)
```

```python
#Define Hyperparameters
np.random.seed(42)
weights = np.random.rand(8,1)
bias = np.random.rand(1)
lr = .0001 #learning rate
weights
```

```python
weights=weights.reshape(8,1)
```

```python
weights
```

```python
bias
```

```python
labels = np.reshape(y, (500,1))
```

```python
#define activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#define derivative
def sigmoid_derivative(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
inputs = X  # set aside the original input
 # dot product of input and add bias
XW = np.dot(inputs, weights)+ bias 
print(XW)
```

```python
z = sigmoid(XW)  #sigmoid activation function
print(z)
error = z - labels  #backprop step1
print(error)
dcost = error
```

```python
dpred = sigmoid_derivative(z)
print(dpred)
```

```python
z_del = dcost * dpred
inputs = X.T
print(z_del)
# print(inputs)
```

```python
print(lr*np.dot(inputs,z_del))
```

```python
# print(inputs[6][:])
```

```python
print(z_del[:][0])
```

```python
weights = weights - lr*np.dot(inputs, z_del)

for num in z_del:
    bias = bias - lr*num
z_del
```

```python
weights
```

```python

#train ANN model
for epoch in range(10000):
    inputs = X  # set aside the original input
    XW = np.dot(inputs, weights)+ bias  # dot product of input and add bias
    z = sigmoid(XW)  #sigmoid activation function
    error = z - labels  #backprop step1
    
    dcost = error
    dpred = sigmoid_derivative(z)

    z_del = dcost * dpred
    inputs = X.T
    
    weights = weights - lr*np.dot(inputs, z_del)
    
    for num in z_del:
        bias = bias - lr*num
```

```python
print(weights)
print(bias)
```

# Predict Points

```python
pred = sigmoid(np.dot(X1, weights) + bias)
pred = pred.astype(int)
```

```python
pred
```

```python
accuracy_score(y1, pred)*100
```
