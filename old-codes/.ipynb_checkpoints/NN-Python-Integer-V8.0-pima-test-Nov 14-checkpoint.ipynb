```python
# taylor series works when x<1
# Assumption based on values of x
# need to check the bwlow formulas of taylor series one more time
# x and y are numerator and denominator
def sigmoidfrac(x,y):
    if(x/y>=10):
        num = 1; den = 1
    elif((x/y)>=9 and (x/y)<10):
        num = 99988; den = 100000
    elif((x/y)>=8 and (x/y)<9):
        num = 99966; den = 100000
    elif((x/y)>=7 and (x/y)<8):
        num = 99909; den = 100000
    elif((x/y)>=6 and (x/y)<7):
        num = 99753; den = 100000
    elif((x/y)>=5 and (x/y)<6):
        num = 99331; den = 100000
    elif((x/y)>=1 and (x/y)<5):
        q = int(x/y) 
        r = x%y
        num= pow(19,q)*24*pow(y,4)
        den= pow(19,q)*24*pow(y,4)+pow(7,q)*(24*pow(y,4)-24*r*pow(y,3)+12*pow(r,2)*pow(y,2)-4*y*pow(r,3)+pow(r,4))    
    elif(x/y>-1 and x/y<1):
        num= 24*pow(y,4)
        den= 48*pow(y,4)-24*x*pow(y,3)+12*pow(x,2)*pow(y,2)-4*y*pow(x,3)+pow(x,4)   
    elif(x/y>-6 and x/y<=-1):
        q = int(x/y) 
        r = x%y
        num = 24*pow(y,4)*pow(7,-q)
        den = 24*pow(y,4)*pow(7,-q)+pow(19,-q)*(24*pow(y,4)+24*r*pow(y,3)+12*pow(r,2)*pow(y,2)+4*y*pow(r,3)+pow(r,4))
    elif(x/y>-7 and x/y<=-6):
        num =15; den= 10000 
    elif(x/y>-8 and x/y<=-7):
        num =6; den = 10000     
    elif(x/y>-9 and x/y<=-8):
        num =2; den = 10000 
    elif(x/y>-10 and x/y<=-9):
         num = 7; den = 100000
    elif(x/y<=-10):
         num = 0 ; den =1
    return num,den
```

```python
# Python Program to find the L.C.M. of two input number
def compute_lcm(x, y):
   # choose the greater number
    if (x > y):
        greater = x
    else:
        greater = y
    while(True):
        if((greater % x == 0) and (greater % y == 0)):
            lcm = greater
            break
        greater += 1
    return lcm
```

```python
# define derivative of numerator and denominator
def sigmoid_derivative(x,y):
    num = x*y-pow(x,2)
    den = pow(y,2)
    return num,den
```

```python
#adding num and den version of the form a/b and c/d
def addition(a,b,c,d):
    den = compute_lcm(b,d)
#     den = b*d
    num = ((den/b)*a) + ((den/d)*c)
#     print(num,den)
    return float(num),float(den)
        
```

```python
def addition_8(x1,x2,x3,x4,x5,x6,x7,x8,y1,y2,y3,y4,y5,y6,y7,y8):
    an12,ad12 = addition(x1,y1,x2,y2)
    an34,ad34 = addition(x3,y3,x4,y4)
    an56,ad56 = addition(x5,y5,x6,y6)
    an78,ad78 = addition(x7,y7,x8,y8)
    an1234,ad1234 = addition(an12,ad12,an34,ad34)
    an5678,ad5678 = addition(an56,ad56,an78,ad78)
    x,y = addition(an1234,ad1234,an5678,ad5678)
    return x,y
```

```python
#subtraction between a/b and c/d
def subtraction(a,b,c,d):
    den = compute_lcm(b,d)
    num = ((den/b)*a) - ((den/d)*c)
    return num,den
```

```python
# multiplying a/b and c/d
def multiply(a,b,c,d): # of the form a/b * c/d
    num = a*c
    den = b*d
    return num,den
```

```python
def bigger(wt):
    n=1000
    for i in range(len(wt)):
        dig = f"{wt[i][0]:.0f}"
        dig = str(dig)
#         print(dig)
        if(len(dig)<n):
            n = len(dig)
    n = n-1
    print(n)
    for i in range(len(wt)):
        dig = f"{wt[i][0]:.0f}"
        dig = str(dig)
        tmp = len(dig)-n
#         print(tmp)
        dig = dig[:tmp]
        dig = int(dig)
        wt[i][0] = dig
        wt[i][1] = wt[i][1]//10000
    return wt
```

```python
def shorter(c,d):
    c = int(c)
    d = int(d)
    if(c<0):
        n = c*(-1)
        if(len(str(n))==len(str(d)) and len(str(n))>2):      
            a = f"{n:.0f}"
            a = str(a[:4])
            a = int(a)*(-1)
            b = f"{d:.0f}"
            b = str(b[:4])
            b = int(b)
#         elif(len(str(d))>=len(str(n))*3):
#             a =int(0)
#             b =int(1)
        elif(len(str(n))>2 and len(str(d))>2):
            a = f"{n:.0f}"
            a = str(a)
            b = f"{d:.0f}"
            b = str(b)
            if(len(str(a))<len(str(b))):
                s = len(str(a))-1
            else:
                s = len(str(b))-1
            a = int(a[:-s])*(-1)
            b = int(b[:-s])
             
        else:
            a = int(n*-1)
            b = d
    elif(len(str(c))==len(str(d)) and len(str(c))>2):      
            a = f"{c:.0f}"
            a = str(a[:4])
            a = int(a)
            b = f"{d:.0f}"
            b = str(b[:4])
            b = int(b)
#     elif(len(str(c))==1 and len(str(d))>3):
#         a = int(0)
#         b = int(1)
    elif(len(str(c))>2 and len(str(d))>2):
            a = f"{c:.0f}"
            a = str(a)
            b = f"{d:.0f}"
            b = str(b)
            if(len(str(a))<len(str(b))):
                s = len(str(a))-1
            else:
                s = len(str(b))-1
            a = int(a[:-s])
            b = int(b[:-s])
    else:
        a = c
        b = d
    return a,b
print(shorter(-4350,5200))
```

```python
def get_exsdxi(inp,res,l):
    inputs = inp.T
    b = np.zeros((inputs.shape[0],2))
    num = 0
    den = 0
    for j in range(inputs.shape[0]):
#         print(inputs[j][:])
        for i in range(res.shape[0]):
            num = num+ inputs[j][i]*res[i][0]
            den = res[i][1]
            
        b[j][0],b[j][1] =shorter(num,den)
        b[j][0]=num
        b[j][1]=den
        num=0
        den=0
    for i in range(b.shape[0]): 
        b[i][0] = b[i][0]*l[0]
        b[i][1] = b[i][1]*l[1]
        b[i][0],b[i][1] =shorter(b[i][0],b[i][1])
    return b
inp_res = get_exsdxi(input_set,res,lr)
print(inp_res)
```

# Get New Weights and Bias

```python
def get_new_weight(a,b):
    c = np.zeros((a.shape[0],2))
    for i in range(len(a)):
        num,den = subtraction(a[i][0], a[i][1], b[i][0],b[i][1])
        c[i][0] = num
        c[i][1] = den
        c[i][0],c[i][1]= shorter(num,den)
    return c
weights = get_new_weight(weights,inp_res)
print(weights)
```
