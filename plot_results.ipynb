```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from numpy import genfromtxt
sig_gas = genfromtxt('sigmoid_gasuse_ganache.csv', delimiter=',')
sig_count = np.arange(0,len(sig_gas),1)
sig_time = genfromtxt('sigmoid_time_ganache.csv', delimiter=',')
sig_count = np.arange(0,len(sig_gas),1)
IXWB_gas =  genfromtxt('IXWB_gasuse_ganache.csv', delimiter=',')
IXWB_count = np.arange(0,len(IXWB_gas),1)
IXWB_time =  genfromtxt('IXWB_time_ganache.csv', delimiter=',')
IXWB_count = np.arange(0,len(IXWB_gas),1)

```

```python

fig, ax = plt.subplots()
ax.plot( sig_count,sig_gas)

ax.set(xlabel='Transaction Number (s)', ylabel='Gas Use (Gwei)',
       title='Sigmoid Function Gas Use')
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.plot( IXWB_count,IXWB_gas)

ax.set(xlabel='Transaction Number (s)', ylabel='Gas Use (Gwei)',
       title='IXWB Function Gas Use')
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.plot( sig_count,sig_time)

ax.set(xlabel='Transaction Number (s)', ylabel='Time (Seconds)',
       title='Sigmoid Function Time')
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.plot( IXWB_count,IXWB_time)

ax.set(xlabel='Transaction Number (s)', ylabel='Time (Seconds)',
       title='IXWB Function Time')
ax.grid()
```
